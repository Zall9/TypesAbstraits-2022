```python
def invar_boucle_ajouter_tas(tab, i, p):
    # Tout chemin de la racine au feuille est bien ordoné a part le chemin qui mene à i

    chemin_vers_i = []
    # On récupere les noeuds au dessus de i (peres)
    j = i//2 # On commence par le pere de i
    while(j != 0): # tant que on est pas aux feuilles
        chemin_vers_i.append(j) # on append le pere dans chemin_vers_i
        j = j // 2 # j deviens le pere 
    k = 0 
    while(k < len(tab)): 
        if(tab[i]>tab[i*2+1] or tab[i]>tab[i*2]): # si l'element i est plus grand que ses fils
            if(k not in chemin_vers_i and k*2 not in chemin_vers_i): # et que ni k ni son fils droit n'est pas un element qui mene vers i
                return False
            if(k not in chemin_vers_i and k*2+1 not in chemin_vers_i): # et que ni k ni son fils gauche n'est pas un element qui mene vers i
                return False
        k = k + 1
    return True
    """
        tout les chemins qui ne contiennent pas le noir sont ordonnés
        plus l'elt monte moins y'a de chemins
        et en dessous tous les chemins qui partent de l'elt noir vers les feuilles (le bas)sont ordonnés

        En sortie de boucle: le père de l'elt noir est plus petit que lui
                => tous les chemins qui sont en dehors sont bon car non touchés & tous les chemins qui passent par lui sont bon également.
            et comme par construction on a un arbre binaire par tas => CQFD
        """
def recupererAllCheminSaufEltI(tabChemin,tas, ind, p, buffer, i):
    """
    Récupère Allchemin (tabChemin est pris dans p buffer)
    
    :param tabChemin: la liste de tous les chemins
    :param tas: le tas
    :param ind: l'index du nœud où nous nous trouvons actuellement
    :param p: L'index du nœud dans le tas
    :param buffer: le chemin actuel
    :return: La liste de tous les chemins possibles de la racine aux feuilles.
    """
    if(ind>p or ind == i): # si l'indice sort du tas ou si l'indice vaut i
        return
    tabChemin.append(buffer) 
    recupererAllCheminSaufEltI(tabChemin,tas, 2*ind, p, buffer + "G", i)
    recupererAllCheminSaufEltI(tabChemin,tas, 2*ind + 1, p,  buffer+"D", i)
def est_feuille(tas,indice,p):
    """
    :param tas: le tas
    :param indice: l'index du nœud où nous nous trouvons actuellement
    :param p: L'index du nœud dans le tas
    :return: True si le nœud est une feuille, False sinon.
    """
    if(indice*2>p and indice*2+1>p): # si le fils gauche et le fils droit sort du tas
        return True
    return False
def recupererAllCheminDessousI(tabChemin,tas, ind, p, buffer):
    """
    Étant donné un tas, un index de nœud et un chemin, la fonction renvoie tous les chemins du nœud I aux
    feuilles du tas
    
    :param tabChemin: la liste de tous les chemins
    :param tas: le tas
    :param ind: l'index du nœud où nous nous trouvons actuellement
    :param p: L'index du nœud dans le tas
    :param buffer: la chaîne qui sera ajoutée au chemin
    :return: La liste de tous les chemins possibles de la racine aux feuilles.
    """
    if(ind > p):
        return
    tabChemin.append(buffer)
    recupererAllCheminDessousI(tabChemin,tas,2*ind,p, buffer + "G")
    recupererAllCheminDessousI(tabChemin,tas,2*ind+1,p, buffer + "D")
def invar_boucle_ajouter_tas2(tab, i, p):
    """
    Étant donné un tas, la fonction renvoie True si le tas est invariant lors de l'insertion d'un nouvel
    élément.
        La fonction renvoie Faux sinon
    :param tab: la table du tas
    :param i: l'index de l'élément à ajouter au tas
    :param p: l'index de l'élément que nous traitons actuellement
    :return: La valeur booléenne de la fonction invar_boucle_ajouter_tas2
    """
    listeCheminsSaufI = []
    listeCheminsSousI = []
    # I est l'élément noir
    recupererAllCheminSaufEltI(listeCheminsSaufI, tab, 1, p, "", i)
    recupererAllCheminDessousI(listeCheminsSousI, tab, i, p, "")
    return bienOrdonne(listeCheminsSaufI, tab, 1) and bienOrdonne(listeCheminsSousI, tab, i)
def bienOrdonne(listeChemins, tab, indDepart):
    """
    Étant donné une liste de chemins, une liste des valeurs des nœuds et l'indice du nœud de départ,
    renvoie True si les valeurs des nœuds sont dans le bon ordre, False sinon
    :param listeChemins: une liste de tous les chemins qui ont été trouvés jusqu'à présent
    :param tab: le tableau représentant le tas
    :param indDepart: l'index du point de départ de la recherche
    :return: une valeur booléenne.
    """
    for chemin in listeChemins:
        indCourant = indDepart
        for i in range(len(chemin)-1):# On parcours les lettre du chemin
            if(chemin[i]=="G"):
                if(tab[indCourant]>tab[indCourant*2]):
                    return False
                indexCourant = indCourant*2
            if(chemin[i]=="D"):
                if(tab[indCourant]>tab[indCourant*2+1]):
                    return False
                indexCourant = indCourant*2 + 1
            
    return True
def estArbreBinaireParfait(tas, p):
    """
    Étant donné un tas, la fonction renvoie True si le tas est un arbre binaire parfait, False sinon
    :param tas: le tas
    :p l'indice tu tableau qui borne le tas contenu dans le tableau
    :return: une valeur booléenne
    """
    h = hauteur(p)
    # On test que chaque niveau sauf le dernier est complet
    for i in range(1,2**h):
        if not isinstance(tas[i], int):
            return False
    
    # on teste si le dernier niveau a des noeuds qui sont groupé le plus a gauche possible
    for i in range(2**h,p): 
        if not isinstance(tas[i], int):
            return False
        
        

    return True
def assertionEstUnTas(tas, p):
    """
    Étant donné un tas, la fonction renvoie True si le tas est bien un tas, False sinon
    
    :param tas: le tas
    :return: La valeur booléenne de la fonction assertionEstUnTas
    """
    listeChemins = []

    # On met i a -1 pour récuperer tout les chemins du tas
    recupererAllCheminSaufEltI(listeChemins,tas, 1, p, "", -1)
    
    return bienOrdonne(listeChemins,tas,1) and estArbreBinaireParfait(tas, p) 
def hauteur(x):
    """Cette fonction donne la hauteur de l'arbre représenté dans un tableau
    Paramètre d'entrée: x => c'est l'indice de la position courrante où l'on est dans le tableau"""
    cpt=0
    while (x!=1):
        x=x//2
        cpt=cpt+1
    return cpt
def invar_boucle_retirer(tas, i, p):
    listeCheminsSaufI = []
    recupererAllCheminSaufEltI(listeCheminsSaufI, tab, 1, p, "", i)
    # Peut etre un and en plus
    return bienOrdonne(listeCheminsSaufI,tas,1) 
def invar_boucle_tri(tas, p, n):
    eSupEgalPplusUn=True
    for e in range (1,p):
        if(tas[e]<=tas[p+1]):
            eSupEgalPplusUn=False
    return assertionEstUnTas(tas,p) and tri_ordre_decroissant(tas, p+1, n) and eSupEgalPplusUn

def conclusion_boucle_tri(tas, p, n):
    return not(p==1 and tas[1]>=tas[2] and tri_ordre_decroissant(tas,2,n)) or tri_ordre_decroissant(tas,1,n)
    
def tri_ordre_decroissant(tas, indDep, indFin):
    for i in range (indDep, indFin):
        if tas[i]<tas[i+1]:
            return False
    return True
```
